```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.DataFrame(np.random.randint(0,100,size=(100, 4)), columns=list('ABCD'))
df = df.to_sparse(fill_value=0)
df.info()
```

```python
sns.jointplot('A', 'B', df)
plt.show()
```

```python
sns.regplot('A', 'B', data=df)
plt.show()
```

```python
sns.regplot('A', 'B', data=df, fit_reg=False)
plt.show()
```

```python
sns.regplot('A', 'B', data=df.to_dense())
plt.show()
```

```python
sns.regplot('A', 'B', data=df.to_dense(), fit_reg=False)
plt.show()
```
